```python
# the summary of data read from the original KDDI dataset

import os
import pandas as pd

# Directories
directory = './ipfix'

# Dictionary to store the number of items in each CSV file
file_counts = {}

for filename in os.listdir(directory):
    filepath = os.path.join(directory, filename)
    # Check if the file is a .gz
    if filename.endswith('.gz'):
        try:
            df = pd.read_csv(filepath, compression='gzip', on_bad_lines='skip', low_memory=False)
            file_counts[filename] = len(df)
        except Exception as e:
            print(f"Error parsing {filename}: {e}")

print(file_counts)
```

```python
# confirm it's 25
print(bool(len(file_counts) == 25))

#just to make x-variable looks nicer
updated_file_counts = {}
for key, value in file_counts.items():
    updated_file_counts[key.split('.')[0]] = value

```

```python
import matplotlib.pyplot as plt
# Create a bar chart with a larger figure size
plt.figure(figsize=(15, 8)) 
plt.bar(updated_file_counts.keys(), updated_file_counts.values())
plt.xlabel('Smart Home Devices')
plt.ylabel('Number of Observations')
plt.title('Number of Observations in 25 Different Smart Home Devices')
plt.xticks(rotation=45, ha='right')  # Rotate labels and align them to the right
plt.tight_layout()  # Adjust layout to make room for the labels
plt.show()
```
